```python
# Step 22. Import the dataset from this address.
# link to data: 'https://raw.githubusercontent.com/aaubs/ds-master/main/data/amazon_prime_titles.csv' & 'https://raw.githubusercontent.com/aaubs/ds-master/main/data/netflix_titles.csv'
# Extra challenge: try downloading it from Kaggle in your Google drive and open it from there
# link to kaggle: https://www.kaggle.com/datasets/shivamb/amazon-prime-movies-and-tv-shows & https://www.kaggle.com/datasets/shivamb/netflix-shows
```

```python
amazon = pd.read_csv("amazon_prime_titles.csv")
netflix = pd.read_csv("netflix_titles.csv")
```

```python
# Step 23. Inspect the new datasets, first lines and the shape
netflix.head(2)
```

```python
amazon.head(2)
```

```python
print('The shape of netflix dataset : ',netflix.shape)
print('The shape of amazon dataset : ',amazon.shape)
print('The shape of disney dataset : ',disney.shape)
```

```python
# Step 24. Concatenante the datasets and create an extra column to say from which platform the values are from
netflix['platform']='netflix'
amazon['platform']='amazon'
disney['platform']= 'disney'
```

```python
df=pd.concat([netflix,amazon,disney],ignore_index=True)
df.info()
```

```python
# Step 25 Check for duplicates
df.duplicated().sum()
```

```python
#Step 26. Check the country that have empty values and remplace it with the mode.
df['country'].fillna(df['country'].mode()[0],inplace=True)
df['country'].isna().value_counts()
```

```python
#Step 27. Check the missing values in the director columns and replace them with "Unknown"
df['director'].fillna('Unknown',inplace=True)
df['director'].isna().value_counts()
```

```python
#Step 28. Do the same for the cast
df['cast'].fillna('Unknown',inplace=True)
df['cast'].isna().value_counts()
```

```python
#Step 29. Drop all missing values.
df=df.dropna()
df.info()
```
